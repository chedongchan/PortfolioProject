```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

#Data Analysis using Python. 
#Skills used: Reading and Manipulating Files with Pandas. Extracting conditional rows and columns. 

#Read CSV DATA using pandas

df = pd.read_csv('PS4_GamesSales.csv', encoding='ANSI')
df.head()
```

```python
#Check Data is Being Read Fully.  
dataframe =  df
print(dataframe)
```

```python
#Check for columns with null values. Display as percentage using numpy.
for col in df.columns:
    missing_pct = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,missing_pct))
```

```python
#Some Columns Have Nulls. Make a new dataframe without them.
df1 = df.dropna()
df1
```

```python
#Check if Nulls are removed.

for col in df1.columns:
    missing_pct = np.mean(df1[col].isnull())
    print('{} - {}%'.format(col,missing_pct))
```

```python
#Nulls removed.
```

```python
#Review data types of each column
df1.dtypes
```

```python
#Year has .0 after all values. Change to INT to remove them.
#For proof-of-concept: Convert salary_in_usd column data type to float65 from int65.
df1['Year'] = df1['Year'].astype('int64')
print(df1)
df1.dtypes
```

```python
#Sort Data by Descending Order of Sales in Europe
sales = df1.sort_values(by = ['Rest of World'], inplace = False, ascending= False)
sales
```

```python
#Assuming there is no 'Global' column, make a column that contains the sales sum from all regions.
sales_1  = sales[['North America', 'Europe','Japan','Rest of World']].sum(axis=1)
sales['TotalSales']=sales_1
sales
```

```python
#Function to extract table data based on Genre. 
def GenreSales(genre):
    sale_table = sales[df['Genre'].isin([str(genre)])]
    return sale_table
```

```python
genre = 'Role-Playing'
```

```python
genre_sales = GenreSales(genre)
```

```python
#Check data without row number limits.
pd.set_option('display.max_rows', None)
genre_sales
```

```python
#Generate a plot of sales across genres
#Config details
plt.title('Sales around the world vs Genre of Game')
plt.ylabel('Sale numbers in ___')
current_values = plt.gca().get_yticks()
plt.ylim(0,20)
plt.xticks(rotation = 90)

plt.scatter(y = sales['TotalSales'], x = sales['Genre'])
```

```python
#Plot statistical test line -  Pearson's correlation from Seaborn Module
sns.regplot(y ='TotalSales', x = 'Year', data = sales, scatter_kws= {"color":"red"}, line_kws= {"color":"blue"} )
corr_data = sales.corr(method='pearson')

plt.show()
```

```python
#Create a heatmap of all correlation data between different columns (and by themselves... always = 1).
sns.heatmap(corr_data, annot= True)
```

```python
#It appears that in general, there is a strong correlation between sales in North America and Europe. If a game sells well in Japn, this does not neessarily translate into success abroad. It appears that either majority of game sales occur in NA and Europe, or success in NA or Europe translates storngly with success across the world. Between years 2013-2020, there is weak to no correlation between success of a game to the release year. 
```
